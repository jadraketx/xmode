```python
import pandas as pd
import numpy as np
```

```python
ping_dtypes = {
    'advertiser_id':'str',
    'platform':'str',
    'location_at':'int',
    'latitude':'float',
    'longitude':'float',
    'altitude':'float',
    'horizontal_accuracy':'float',
    'vertical_accuracy':'float',
    'heading':'float',
    'speed':'float',
    'ipv_4':'str',
    'ipv_6':'str',
    'final_country':'str',
    'user_agent':'str',
    'background':'str',
    'publisher_id':'str',
    'wifi_ssid':'str',
    'wifi_bssid':'str',
    'venue_name':'str',
    'dwell_time':'str'
}
```

```python
dat = pd.read_csv(
    "../filtered//part-00001-b6931cd9-5e48-47fe-adf0-70f983cd4630-c000.csv.gz",
    usecols = [e for e in ping_dtypes],
    dtype = ping_dtypes,
    escapechar="\\",
)
```

```python
#string formatting
dat['wifi_ssid'] = dat['wifi_ssid'].str.replace('"','')
dat['wifi_ssid'] = dat['wifi_ssid'].str.strip()
dat['wifi_ssid'] = dat['wifi_ssid'].replace('<unknown ssid>',np.nan)
dat['wifi_bssid'] = dat['wifi_bssid'].replace('<unknown bssid',np.nan)
```
